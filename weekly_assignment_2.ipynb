```python
# Import all the neccesary libraries
import seaborn as sns #this is the plotting library I'll be using 
import pandas as pd #"as pd" means that we can use the abbreviation in commands
import matplotlib.pyplot as plt #we need Matplotlib for setting the labels in the Seaborn graphs

df1 = pd.read_csv("steps.csv", sep=";")
df2 = pd.read_csv("survey.csv")
df1.head()
```

```python
df2.head()
```

```python
# Calculate the average steps per user and create new variable Steps_average
col = df1.loc[: , "20-6-2013":"13-5-2014"]
df1["mean steps"] = col.mean(axis=1)
df1.head()
```

```python
# Merge two csv files
df = pd.merge(df1, df2, on = 'id') #'id' is the common identifier
df.head()
```

The relation of meansteps per participant with at least 2 other quantitative variables

```python
print (df.info()) # Get information on the variables
```

Steps < 30000 (around 25 km)
Height < 190 
Weight < 100

```python
df = df[(df["mean steps"] < 30000) & (df["height"] < 190) & (df["weight"] < 100)] # Subset the original dataset
df.head()
```

```python
df_subset = df[["mean steps", "height", "weight"]] # Subset the original dataset
df_subset.head()
```

```python
print (df_subset.info()) # Get information on the variables
```

```python
sns.pairplot(df_subset)
plt.show() # Use Matplotlib to show the plot
```

```python
sns.lmplot(x = "weight", y = "mean steps", fit_reg=False, data=df) #fit_reg=False leaves out the regression line
plt.xlabel("Weight") # Use matplotlib to refine Seaborn plot
plt.ylabel("Average steps")
plt.title("Average steps and weight")
plt.show()
```

There is no linear relationship between as the points do not lie around a virtual line. There is a downhill pattern which indicates a negative relationship between height and average steps.

```python
sns.lmplot(x = "height", y = "mean steps", fit_reg=False, data=df) #fit_reg=False leaves out the regression line
plt.xlabel("Height") # Use matplotlib to refine Seaborn plot
plt.ylabel("Average steps")
plt.title("Average steps and height")
plt.show()
```

There is no linear relationship between as the points do not lie around a virtual line. There is a downhill pattern which indicates a negative relationship between height and average steps.

```python
from scipy.stats.stats import pearsonr # Pearson's corelation
print(pearsonr(df_subset["mean steps"],df["weight"]))
print(pearsonr(df_subset["mean steps"],df["height"]))
```

Based on r which is -0.08, and p which appears to be 0.01 we can conclude the following:

Average steps and weight are negatively correlated (r = -0.08) as the value is very close to zero, the correlation is very low.
The correlation is significant (p = 0.01 which means p < 0.05)

Average steps and height are negatively correlated (r = -0.08)as the value is very close to zero, the correlation is very low.
The correlation is significant (p = 0.007, which means p < 0.05)

```python
corr = df_subset[["mean steps", "weight", "height"]].corr() # To calculate all the correlations and make a correlation matrix 
corr
```

```python
sns.heatmap(corr, xticklabels=corr.columns, yticklabels=corr.columns)# Create correlation matrix
plt.show()
```
